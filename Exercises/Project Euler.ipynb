The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
longest = 0
found_it=0
for i in range (800001,1000001,2):
    number=i
    
    operation=[]

    while number != 1 :
        operation.append(number)
        if number%2==0:
            number=number/2
        else:
            number=3*number+1


    operation.append(1)

    if len(operation)>longest:
        longest = len(operation)
        found_it=i


print(f"{found_it} has {longest}")
    
        

```

```python
import math

def divisors_of(n):
    large_divisors = [1]
    for i in range(2, int(math.sqrt(n) + 1)):
        if n % i == 0:
            yield i
            if i*i != n :
                large_divisors.append(int(n / i))
    for divisor in reversed(large_divisors):
        yield divisor

def type_of(n):
    if n==sum(list(divisors_of(n))):
        return "Perfect"
    elif n>sum(list(divisors_of(n))):
        return "Deficient"
    else:
        return "Abundant"



limit = 28124
abundant_numbers=[]


for i in range(1,limit+1,1):
    if type_of(i)=="Abundant":
        abundant_numbers.append(i)


all_additions=[]
for i in range(0,len(abundant_numbers),1):
    A=abundant_numbers[i]
    for j in range(i,len(abundant_numbers),1):
        B=abundant_numbers[j]
        if A+B > limit :
            break
        if A+B not in all_additions :
            all_additions.append(A+B)


not_addditionnable=[0]
for i in range(0,limit+1,1):
    if i not in all_additions:
        not_addditionnable.append(i)

    

print(sum(not_addditionnable))



    




```

012   021   102   120   201   210


# Problem 27: Quadratic primes

```python
def is_prime(x):
    if x==0:
        return False
    elif x==1:
        return False
    elif x==2:
        return True
    elif x%2==0:
        return False
    else: 
        for i in range(3,int(x/2 + 1),2):
            if x%i==0:
                return False
        return True


limit=1000
longest=0


for i in range (1-limit,limit,1):
    for j in range(1-limit,limit,1):
        n=0
        in_a_row=0

        if is_prime(n**2+i*n+j):
            while is_prime(n**2+i*n+j):
                n+=1
                in_a_row+=1

        if in_a_row>longest:
            longest=in_a_row
            a=i
            b=j
            print(f"{longest} consecutive prime for a={a} and b={b}")

print(longest)
print(a)
print(b)
print(a*b)
            
```
